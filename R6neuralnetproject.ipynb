```
import pandas as pd
```

```
project_path = "/content/drive/My Drive/Colab Notebooks/"
```

```
data = pd.read_csv('bank.csv')
```

```
#Drop the columns which are unique for all users like IDs (2.5 points)
```

```
data.head()
```

```
data['RowNumber']
#removing row number as it is same as index
```

```
data.drop(columns=['CustomerId','RowNumber','Surname'],inplace=True)
```

```
data = pd.get_dummies(data)
```

```
X = data.drop(columns=['Exited'])
y = data['Exited']
```

```
# Normalize the train and test data (2.5 points)
#Divide the data set into Train and test sets
```

```
#import sklean
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import Normalizer
```

```
norm = Normalizer()
X_norm = norm.fit_transform(X)
```

```
X_train,X_test,y_train,y_test = train_test_split(X_norm,y,test_size=0.25,random_state=123)
```

```
#Initialize & build the model (10 points)
```

```
import tensorflow as tf
import keras
tf.set_random_seed(42)

```

```
model = tf.keras.Sequential()
```

```
model.add(tf.keras.layers.Dense(2,activation='softmax'))
```

```
model.compile(optimizer='sgd',loss='categorical_crossentropy',metrics=['accuracy'])
```

```
y_test = tf.keras.utils.to_categorical(y_test)
y_train = tf.keras.utils.to_categorical(y_train)
```

```
model.fit(X_train,y_train,epochs=30)
```

```
#Optimize the model (5 points)
```

```
model2 = tf.keras.Sequential()
model2.add(tf.keras.layers.BatchNormalization())
model2.add(tf.keras.layers.Dense(2,activation='softmax'))
model2.compile(optimizer='sgd',loss='categorical_crossentropy',metrics=['accuracy'])
```

```
model2.fit(X_train,y_train,epochs=30,batch_size=30)
```

```
model4 = tf.keras.Sequential()
model4.add(tf.keras.layers.BatchNormalization())
model4.add(tf.keras.layers.Dense(100,activation='sigmoid'))
model4.add(tf.keras.layers.Dense(100,activation='sigmoid'))
model4.add(tf.keras.layers.Dense(2,activation='softmax'))
model4.compile(optimizer='sgd',loss='categorical_crossentropy',metrics=['accuracy'])
```

```
model4.fit(X_train,y_train,epochs=30,batch_size=30)
```

```
pred = model4.predict(X_test)
```

```
pred_df = pd.DataFrame(columns=['y_test','y_pred'])
```

```
y_pred =[]
for i in range(pred.shape[0]):
  if pred[i][0] >= 0.5:
    pred_df.loc[i] = [y_test[i][1]]+[0.0]
    
  else:
    pred_df.loc[i] = [y_test[i][0]]+[1.0]
```

```
pred_df.head()
```

```
from sklearn.metrics import confusion_matrix,accuracy_score
```

```
confusion_matrix(pred_df['y_test'],pred_df['y_pred'])
```

```
accuracy_score(pred_df['y_test'],pred_df['y_pred'])
```
