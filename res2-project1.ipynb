### Q2. Use the five percent significance test over the data to determine the p value to check new scheme has significantly raised outputs? (10 points)

```python
old=np.array(employee['Old Scheme (in thousands)'])
new=np.array(employee['New Scheme (in thousands)'])
```

```python
old
```

```python
new
```

```python
''' µ1 be the mean of old scheme , µ2 be the mean of new scheme
    As we are comparing the values of same experimental unit with two different values,this is a scenario of paired t-test
    we will consider H0 as µ1-µ2<=0 instead of µ1=µ2 
    H1 : µ2-µ1>0 '''
#Basic assumption for t-test is the data is normally distributed , variances between the groups are equal. 
#Using shapiro we'll confirm whether it is normally dist or not
from scipy.stats import shapiro,levene
print("old scheme: {}".format(shapiro(old)))
print("new scheme : {}".format(shapiro(new)))



    
```

```python
#As P value is very high , we can confirm that given data is normally distrbuted
#now let's check for equality of variances
levene(old,new)
```

```python
#For 95% confidence level ,as pvalue >0.05 , we can confirm that the variances are equal for old,new data 
```

```python
#performing paired t-test
from scipy.stats import ttest_1samp
tstat,pval=ttest_1samp(new-old,0)
#as we are considering only raise in output i.e µ2-µ1 >0 , we consider only right tail
print("P-Value of paired t-test is {}".format(round(pval/2,3)))
```

#### Q3. What conclusion does the test (p-value) lead to? (2.5 points)

```python
#for 95% significance ,as p value = 0.065 > 0.05 , H0 should not be rejected.
# µ1-µ2 = 0  i.e µ2 = µ1 .
'''There is NO signficant raise in the output of new scheme compared to old scheme'''
```

```python
for i in range(11):
    t,p1=ttest_1samp(new-old,i)
    print("P-Val for diff of {} is {} : We are {}% confident to reject H0 but not 95% ".format(i,round(p1,2),round(1-p1,2)))
```

```python
# from 0-4 our confidence of rejecting decreases so it is clear that diff is some where around 4
# 0-4 covers µ2-µ1 <4
# 4-9 covers µ2-µ1 >4
# from 10 , diff is so large that we can reject
# Hence for 95% confidence  -1<=µ2-µ1<=9
```
